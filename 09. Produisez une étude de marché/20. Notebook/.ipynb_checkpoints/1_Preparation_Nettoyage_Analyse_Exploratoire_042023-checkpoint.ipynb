```python
def Afficher_Données_Colonnes(dataframe):
    for colonne in dataframe.columns:
        print("La Colonne '{}' contient les valeurs suivantes: {}".format(colonne, dataframe[colonne].unique()), "\n")
```

<div style="background-color:  RGB(255,224,132);" >
<h3 style="margin: auto; padding: 20px; color:#861D00; "> Comparaison de deux colonnes </h3>
</div>

```python
# Comparer deux colonnes d'un même dataframe
def Comparaison_Colonne(dataframe, colonne1, colonne2):
    if dataframe[colonne1].equals(dataframe[colonne2]):
        return print("Les données de la colonne", colonne1, "et", colonne2, "sont identiques")
    else:
        return print("Les données de la colonne", colonne1, "et", colonne2, " ne sont pas identiques")
```

<div style="background-color: RGB(255,224,132);" >
<h3 style="margin: auto; padding: 20px; color:#861D00; "> Contrôle des Nan </h3>
</div>

```python
def ControleNan(dataframe):
    ComptageNan = dataframe.isna().sum().sum()
    if ComptageNan > 0:
        print("Il y a {} valeurs NaN dans le dataframe".format(ComptageNan))
        LigneNan = dataframe[dataframe.isna().any(axis=1)]
        return LigneNan
    else:
        print("Il n'y a pas de valeurs NaN dans le dataframe.")
        LigneNan = dataframe[dataframe.isna().any(axis=1)]
        return LigneNan
```

```python
# Récupérer les données de la Population 
donneesFinales = Population_Finale.copy()
# Jointure avec le fichier ImportVolailles
donneesFinales = panda.merge(donneesFinales, ImportVolailles, how="left", on='code zone')
# Jointure avec le fichier TauxVolaillesViandes
donneesFinales = panda.merge(donneesFinales, TauxVolaillesViandes, how="left", on='code zone')
# Jointure avec le fichierTauxViandesNourriture
donneesFinales = panda.merge(donneesFinales, TauxViandesNourriture, how="left", on='code zone')
# Jointure avec le fichier SousAlimentation
donneesFinales = panda.merge(donneesFinales,SousAlimentation, how="left", on='code zone')
# Jointure avec le fichier PIB
donneesFinales = panda.merge(donneesFinales, Pib, how="left", on='code zone')
# Jointure avec le fichier Stabilité Politique
donneesFinales = panda.merge(donneesFinales, StabilitePolitique, how="left", on='code zone')
# Jointure avec le fichier Facilité de faire des affaires
donneesFinales = panda.merge(donneesFinales, FaciliteAffaire, how="left", on='code zone')
# Jointure avec le fichier distance
donneesFinales = panda.merge(donneesFinales, Distance, how="left", on='code zone')
# Afficher les données
donneesFinales.to_excel(repertoireFichierImportation + 'donneesFinales.xlsx', sheet_name='donneesFinales', index=False)
donneesFinales
```
